```python
import re
import folium
import pandas as pd

# to load in the function from dataverzameling/combination.py to extract lat and lon
import importlib.util
spec = importlib.util.spec_from_file_location("combination", "../dataverzameling/combination.py")
combination = importlib.util.module_from_spec(spec)
spec.loader.exec_module(combination)
extract_latitude_longitude = combination.extract_latitude_longitude


filename = "coordinates.txt"
data = []

with open(filename, 'r') as file:
    log_entries = file.readlines()
    
coordinates = []

for i in range(1750, len(log_entries)):
    lat, lon = extract_latitude_longitude(log_entries[i])
    coordinates.append((float(lat), float(lon)))

df = pd.DataFrame(coordinates, columns=["latitude", "longitude"])

# Create a map centered around a specific location (replace with your desired location)
m = folium.Map(location=[df['latitude'].mean(), df['longitude'].mean()], max_zoom=30, zoom_start=17, tiles='https://mt1.google.com/vt/lyrs=s&x={x}&y={y}&z={z}', attr='Google')

for idx, row in df.iterrows():
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=1,  # this controls the size of the markers
        color='red',
        fill=True,
        fill_color='red',
        fill_opacity=0.6,
        popup=str(idx),
        tooltip=str(idx)  # add the ID as a tooltip label
    ).add_to(m)

# Display the map
m
```
